En la segunda parte de nuestro taller realizaremos un ajuste de datos por medio de mínimos cuadrados para diferentes periodos de tiempo.  
  
  El primer lapso está comprendido entre 1980 y 2000.

```python
from numpy import loadtxt

dataCO2 = loadtxt("co2_mm_gl.txt", float)

plot(dataCO2[:,2], dataCO2[:,3])
xlim(1980, 2000)
ylim(335, 370)
show()
```

```python
from numpy import loadtxt
import matplotlib.pyplot as plt
import numpy as np

dataCO2 = loadtxt("co2_mm_gl.txt", float)
x = dataCO2[:,2]
y = dataCO2[:,3]
media_x = x.mean()
media_y = y.mean()

# Hallamos la pendiente 
m = np.sum((x - media_x)*(y - media_y)) / np.sum((x - media_x)**2)

# Hallamos la intercepción
b = media_y - m * media_x

plot(dataCO2[:,2], dataCO2[:,3],color="green")
xlim(1980, 2000)
ylim(335, 370)
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = b + m * x_vals
plt.plot(x_vals, y_vals, color="black")
plt.show()
print ("La pendiente es: ",m)
print ("El intercepto es: ",b)
print ("La función que describe el ajuste lineal es: 1.7794x-3188.5536")
rss = np.sum((y - (b + m * x)) ** 2)
print ("El error del ajuste es: ",rss)
```
